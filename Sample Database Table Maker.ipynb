```python
import pandas as pd #Pandas is used to create our tables
import dataframe_image as dfi #Dataframe_images lets us save these tables as pngs
```

```python
Samples = pd.read_excel('Minerals_of_Rhode_Island_List.xlsx') #This reads the excel file and names it Samples
dfi.export(Samples, 'Minerals_of_Rhode_Island_raw.png') #This takes the dataframe and saves it as a png
Samples
```

```python
#Here we sort our samples alphabetically by each column and set the sorting to it's own name 
Sort_Name=Samples.sort_values('Name')
Sort_Chemical=Samples.sort_values('Chemical Composition')
Sort_Origin=Samples.sort_values('Location of Origin')
Sort_Apperance=Samples.sort_values('Primary Apperance')
```

# Now we can choose which sorting we want for our table
### Each block indexs the table by a different column which is sorted alphabetically then saves the table as a png

```python
#Index and sort by name
SoNai=Sort_Name.set_index('Name')
dfi.export(SoNai, 'Minerals_of_Rhode_Island_name.png')
```

```python
#Index and sort by chemcial composition
SoChi=Sort_Chemical.set_index('Chemical Composition')
dfi.export(SoChi, 'Minerals_of_Rhode_Island_chemical.png')
```

```python
#Index and sort by location of origin
SoOri=Sort_Origin.set_index('Location of Origin')
dfi.export(SoOri, 'Minerals_of_Rhode_Island_origin.png')
```

```python
#Index and sort by primary apperance
SoApi=Sort_Apperance.set_index('Primary Apperance')
dfi.export(SoApi, 'Minerals_of_Rhode_Island_apperance.png')
```
